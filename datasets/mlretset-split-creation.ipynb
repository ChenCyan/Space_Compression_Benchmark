```python
import glob
import numpy as np
import random

random.seed(10587)
```

```python
in_directory = "./MLRetSet/data/"

out_directory = "./MLRetSet/splits/"
```

```python
train_percent = 0.7
val_percent = 0.2
test_percent = 0.1
```

```python
in_patches = glob.glob(f"{in_directory}/*.mat")
in_patches = [("/").join(patch.split("/")[-1:]) for patch in in_patches]

random.shuffle(in_patches)
```

```python
num_patches = len(in_patches)
```

```python
train_patches_easy, val_patches_easy, test_patches_easy = np.split(in_patches, [int(num_patches*train_percent), int(num_patches*(train_percent+val_percent)+1)])

train_patches_easy = list(train_patches_easy)
val_patches_easy = list(val_patches_easy)
test_patches_easy = list(test_patches_easy)
```

```python
train_patches_easy.sort()
val_patches_easy.sort()
test_patches_easy.sort()
```

```python
np.savetxt(f"{out_directory}/train.csv", train_patches_easy, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/val.csv", val_patches_easy, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/test.csv", test_patches_easy, delimiter =", ", fmt ='% s')
```
