```python
import glob
import numpy as np
import random

random.seed(10587)
```

```python
in_directory = "./berlin-urban-gradient/patches/"

out_directory = "./berlin-urban-gradient/splits/"
```

```python
train_percent = 0.7
val_percent = 0.2
test_percent = 0.1
```

```python
in_patches = glob.glob(f"{in_directory}/*.npy")
in_patches = [("/").join(patch.split("/")[-1:]) for patch in in_patches]

random.shuffle(in_patches)
```

```python
num_patches = len(in_patches)
```

```python
train_patches, val_patches, test_patches = np.split(in_patches, [int(num_patches*train_percent), int(num_patches*(train_percent+val_percent))])

train_patches_easy = list(train_patches)
val_patches_easy = list(val_patches)
test_patches_easy = list(test_patches)
```

```python
train_patches.sort()
val_patches.sort()
test_patches.sort()
```

```python
np.savetxt(f"{out_directory}/train.csv", train_patches, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/val.csv", val_patches, delimiter =", ", fmt ='% s')
np.savetxt(f"{out_directory}/test.csv", test_patches, delimiter =", ", fmt ='% s')
```
